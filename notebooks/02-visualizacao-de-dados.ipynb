# 1. Visualização de dados

```python
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt
```

## 1.1. Leitura dos dados

```python
df = pd.read_csv('https://gist.githubusercontent.com/omadson/770934cb0df9e2d2c3e8db031f1139a5/raw/4d181efc035b327d9d66256915951384517c603e/arrecadacao.csv')
df.head()
```

## 1.2. Diagrama de Barras Verticais

```python
xy = sns.barplot(data = df, y = 'arrecadacao', x = 'titulo')
```

## 1.3. Diagrama de Barras Horizontais

```python
df = df.assign(arrecadacao_milhoes = df.arrecadacao / 1000000)
df = df.sort_values('arrecadacao_milhoes', ascending = False)

sns.set_style("whitegrid")
plt.figure(figsize = (5,2))
bar = sns.barplot(data = df, x = 'arrecadacao_milhoes', y = 'titulo', color = 'teal')

bar.bar_label(
    bar.containers[0],
    fmt='%1.1f',
    label_type='center',
    color='white'
)

plt.xlabel('Arrecadação (US$ Milhões)')
plt.ylabel('Título')
plt.title('Arrecadação dos Filmes')

bar.set(frame_on = False)
bar.axvline(0, color = "k", clip_on = False)
plt.show()
```
